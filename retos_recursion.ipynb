# Práctica: Retos de Recursión



## ¿Cómo realizar esta práctica?

Cada reto incluye:

- 🎯 **Objetivo**
- 🕒 **Tiempo estimado**
- 🟢🟡🔴 **Dificultad**
- 🧠 Preguntas de diseño
- 💡 Pistas
- ✅ Pruebas automáticas con `assert`

Antes de escribir código, responde mentalmente:

1. ¿Cuál es el caso base?
2. ¿Cómo reduzco el problema?
3. ¿Cada llamada se acerca al caso base?
4. ¿Qué valor devuelve cada llamada?

> ## 📌 Reglas de la práctica
>
> Salvo que el enunciado indique lo contrario:
>
> - Las soluciones deben ser **recursivas**.
> - No se permite utilizar bucles `for` o `while`.
> - No se permite utilizar funciones que resuelvan directamente el problema, como `sum`, `max`, `reversed` o `sorted`.
> - Puedes crear funciones auxiliares cuando lo necesites.
