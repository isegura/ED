## 1. Cadenas de texto

Las cadenas de texto, o `str`, sirven para guardar texto.

Ejemplos:
- nombres
- mensajes
- ciudades
- correos electrónicos

Se escriben entre comillas simples `' '` o dobles `" "`.

### ¿Para qué sirven?

Sirven para representar y manipular información textual.


````python
# Ejemplo básico con cadenas

nombre = "Lucía"
apellido = "Gómez"

print("Nombre completo:", nombre + " " + apellido)
print("Número de letras del nombre:", len(nombre))
print("Nombre en mayúsculas:", nombre.upper())
````

## 3. Listas

Las listas permiten guardar varios valores en una sola variable.

Ejemplo:
```python
notas = [7, 8, 5, 9]
```

Las listas:
- mantienen el orden
- pueden modificarse
- admiten elementos repetidos

### ¿Para qué sirven?

Sirven para almacenar colecciones de datos, por ejemplo:
- notas de un alumno
- nombres de una clase
- precios de productos


````python
# Ejemplo básico con listas

notas = [7, 8, 5, 9]

print("Lista de notas:", notas)
print("Primera nota:", notas[0])

notas.append(10)
print("Lista después de añadir una nota:", notas)
print("Número total de notas:", len(notas))
````

## 4. Tuplas

Las tuplas son parecidas a las listas, pero no se pueden modificar después de crearse.

Ejemplo:
```python
coordenadas = (3, 5)
```

### ¿Para qué sirven?

Sirven para guardar datos que no deben cambiar, por ejemplo:
- coordenadas
- fechas
- datos fijos


````python
# Ejemplo básico con tuplas

coordenadas = (10, 20)

print("Coordenadas:", coordenadas)
print("Valor X:", coordenadas[0])
print("Valor Y:", coordenadas[1])
````

## 5. Diccionarios

Los diccionarios guardan información en pares `clave: valor`.

Ejemplo:
```python
alumno = {"nombre": "Ana", "edad": 15}
```

### ¿Para qué sirven?

Sirven para representar datos relacionados, por ejemplo:
- información de una persona
- datos de un producto
- configuración de una aplicación


````python
# Ejemplo básico con diccionarios

alumno = {
    "nombre": "Ana",
    "edad": 15,
    "curso": "4º ESO"
}

print("Diccionario completo:", alumno)
print("Nombre:", alumno["nombre"])
print("Curso:", alumno["curso"])

alumno["edad"] = 16
print("Edad actualizada:", alumno["edad"])
````

## 6. Conjuntos

Los conjuntos (`set`) almacenan elementos sin orden y sin repetidos.

Ejemplo:
```python
numeros = {1, 2, 3, 4}
```

### ¿Para qué sirven?

Sirven para:
- eliminar duplicados
- comprobar pertenencia rápidamente
- trabajar con grupos de elementos únicos


````python
# Ejemplo básico con conjuntos

numeros = {1, 2, 2, 3, 4, 4, 5}

print("Conjunto:", numeros)

numeros.add(6)
print("Después de añadir 6:", numeros)

print("¿Está el 3 en el conjunto?", 3 in numeros)
````
