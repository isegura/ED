# 🏆 Reto 6. Implementando el TAD Marcador

🎯 **Objetivo:** Implementar un Tipo Abstracto de Datos utilizando una clase de Python.

🕒 **Tiempo estimado:** 20 minutos

🟡 **Dificultad:** Media

### 📋 Enunciado

En el reto anterior has identificado los atributos y métodos del **TAD Marcador**, que representa el resultado de un partido de fútbol.

Ahora implementa la clase `Marcador` en Python.

La clase debe incluir:

- Un constructor que reciba el número de goles del equipo local y del equipo visitante.
- Los atributos:
  - `goles_local`
  - `goles_visitante`
- Un método `gol_local()` que incremente en uno los goles del equipo local.
- Un método `gol_visitante()` que incremente en uno los goles del equipo visitante.
- Un método `resultado()` que devuelva una cadena con el resultado en el formato:

```text
goles_local - goles_visitante
```

Por ejemplo:

```python
marcador = Marcador(1, 0)

marcador.gol_local()
marcador.gol_visitante()
marcador.gol_visitante()

print(marcador.resultado())
```

Salida esperada:

```text
2 - 2
```

### 🧠 Antes de empezar

- Recuerda que el constructor debe inicializar el estado del objeto.
- Los métodos `gol_local()` y `gol_visitante()` modifican ese estado.
- El método `resultado()` no modifica el objeto; únicamente devuelve una representación del marcador.

````python
class Marcador:
    """Representa el resultado de un partido de fútbol."""

    def __init__(self, goles_local, goles_visitante):
        self.goles_local = goles_local
        self.goles_visitante = goles_visitante

    def gol_local(self):
        self.goles_local += 1

    def gol_visitante(self):
        self.goles_visitante += 1

    def resultado(self):
        return f"{self.goles_local} - {self.goles_visitante}"
````

Ejecuta la siguiente celda para comprobar tu solución.

````python
marcador = Marcador(1, 0)

assert marcador.goles_local == 1
assert marcador.goles_visitante == 0

marcador.gol_local()
assert marcador.goles_local == 2
assert marcador.goles_visitante == 0

marcador.gol_visitante()
marcador.gol_visitante()

assert marcador.goles_local == 2
assert marcador.goles_visitante == 2

assert marcador.resultado() == "2 - 2"

print("¡Pruebas superadas!")
````

# 🏆 Reto 8. Encapsulación

🎯 **Objetivo:** Implementa una clase `Termostato` que controle que la temperatura permanezca en un intervalo válido.

🕒 **Tiempo estimado:** 15 minutos

🟡 **Dificultad:** Media

La clase debe incluir:

- atributo `temperatura`
- método `establecer_temperatura(nueva)`
- método `consultar_temperatura()`

Solo se admitirán temperaturas entre `5` y `35` grados, ambos incluidos.
`establecer_temperatura` devolverá `True` si modifica el valor y `False` si lo rechaza.

### 🧠 Antes de programar
- La validación debe estar dentro del método.
- Si el valor es inválido, el estado del objeto debe conservarse.

````python
class Termostato:
    """Mantiene una temperatura dentro del intervalo permitido."""

    def __init__(self, temperatura):
        self.temperatura = temperatura

    def establecer_temperatura(self, nueva):
        if 5 <= nueva <= 35:
            self.temperatura = nueva
            return True
        return False

    def consultar_temperatura(self):
        return self.temperatura
````

Ejecuta la siguiente celda para comprobar tu solución.

````python
termostato = Termostato(20)

assert termostato.consultar_temperatura() == 20
assert termostato.establecer_temperatura(25) is True
assert termostato.consultar_temperatura() == 25

assert termostato.establecer_temperatura(40) is False
assert termostato.consultar_temperatura() == 25

assert termostato.establecer_temperatura(5) is True
assert termostato.establecer_temperatura(35) is True
print("Pruebas superadas")
````

# 🏆 Reto 9. Herencia

🎯 **Objetivo:** Completa una jerarquía sencilla formada por `Animal`, `Perro` y `Gato`.

🕒 **Tiempo estimado:** 12 minutos

🟡 **Dificultad:** Media

La clase base `Animal` almacenará `nombre`.

Cada subclase debe implementar `hacer_sonido()`:

- `Perro` devuelve `"guau"`
- `Gato` devuelve `"miau"`

### 🧠 Antes de programar
- ¿Qué atributo comparten las subclases?
- Utiliza `super()` para reutilizar el constructor de la clase base.

````python
class Animal:

    def __init__(self, nombre):
        self.nombre = nombre


class Perro(Animal):

    def __init__(self, nombre):
        super().__init__(nombre)

    def hacer_sonido(self):
        return "guau"


class Gato(Animal):

    def __init__(self, nombre):
        super().__init__(nombre)

    def hacer_sonido(self):
        return "miau"
````

Ejecuta la siguiente celda para comprobar tu solución.

````python
perro = Perro("Luna")
gato = Gato("Milo")

assert perro.nombre == "Luna"
assert gato.nombre == "Milo"
assert perro.hacer_sonido() == "guau"
assert gato.hacer_sonido() == "miau"
assert isinstance(perro, Animal)
assert isinstance(gato, Animal)
print("Pruebas superadas")
````

# 🏆 Reto 10. Polimorfismo

🎯 **Objetivo:** Implementa una función que trabaje con distintos objetos que comparten una misma operación.

🕒 **Tiempo estimado:** 10 minutos

🟡 **Dificultad:** Media

Implementa `sonidos(animales)`.

La función recibe una lista de objetos y devuelve una lista con el resultado de llamar a `hacer_sonido()` en cada uno.

### 🧠 Antes de programar
- La función no necesita comprobar si cada objeto es un perro o un gato.
- Basta con que todos ofrezcan el mismo método.

````python
def sonidos(animales):
    """Devuelve el sonido producido por cada objeto de la colección."""
    return [animal.hacer_sonido() for animal in animales]
````

Ejecuta la siguiente celda para comprobar tu solución.

````python
class Perro:
    def hacer_sonido(self):
        return "guau"

class Gato:
    def hacer_sonido(self):
        return "miau"

class Pajaro:
    def hacer_sonido(self):
        return "pio"

animales = [Perro(), Gato(), Pajaro()]
assert sonidos(animales) == ["guau", "miau", "pio"]
assert sonidos([]) == []
print("Pruebas superadas")
````

# 🏆 Reto 11. Depurar una clase

🎯 **Objetivo:** Corrige los errores de la clase `Producto`.

🕒 **Tiempo estimado:** 12 minutos

🟡 **Dificultad:** Media

La clase debe:

- almacenar `nombre` y `precio`
- permitir aplicar un descuento porcentual
- rechazar descuentos menores que `0` o mayores que `100`
- devolver `True` cuando se aplica el descuento y `False` cuando se rechaza

### 🧠 Antes de programar
Busca errores relacionados con:

- uso de `self`
- modificación de atributos
- validación de datos

````python
class Producto:

    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def aplicar_descuento(self, porcentaje):
        if 0 <= porcentaje <= 100:
            self.precio -= self.precio * porcentaje / 100
            return True
        return False
````

Ejecuta la siguiente celda para comprobar tu solución.

````python
producto = Producto("Teclado", 80)

assert producto.nombre == "Teclado"
assert producto.precio == 80

assert producto.aplicar_descuento(25) is True
assert producto.precio == 60

assert producto.aplicar_descuento(-10) is False
assert producto.precio == 60

assert producto.aplicar_descuento(120) is False
assert producto.precio == 60
print("Pruebas superadas")
````

# 🏆 Reto 12. Implementar el TAD Número Complejo

🎯 **Objetivo:** Implementa una clase sencilla para representar números complejos.

🕒 **Tiempo estimado:** 20 minutos

🔴 **Dificultad:** Difícil

La clase `NumeroComplejo` debe almacenar:

- `real`
- `imaginaria`

Debe implementar:

- `sumar(otro)`, que devuelve un nuevo `NumeroComplejo`
- `restar(otro)`, que devuelve un nuevo `NumeroComplejo`
- `modulo()`, que devuelve su módulo

### 🧠 Antes de programar
Para un número complejo `a + bi`:

- la suma y la resta se realizan componente a componente;
- el módulo es `(a² + b²) ** 0.5`.

````python
class NumeroComplejo:

    def __init__(self, real, imaginaria):
        self.real = real
        self.imaginaria = imaginaria

    def sumar(self, otro):
        return NumeroComplejo(
            self.real + otro.real,
            self.imaginaria + otro.imaginaria
        )

    def restar(self, otro):
        return NumeroComplejo(
            self.real - otro.real,
            self.imaginaria - otro.imaginaria
        )

    def modulo(self):
        return (self.real ** 2 + self.imaginaria ** 2) ** 0.5
````

Ejecuta la siguiente celda para comprobar tu solución.

````python
z1 = NumeroComplejo(3, 4)
z2 = NumeroComplejo(1, -2)

suma = z1.sumar(z2)
resta = z1.restar(z2)

assert isinstance(suma, NumeroComplejo)
assert suma.real == 4
assert suma.imaginaria == 2

assert resta.real == 2
assert resta.imaginaria == 6

assert z1.modulo() == 5
print("Pruebas superadas")
````

# 🏆 Reto 13. Modelado de una biblioteca

🎯 **Objetivo:** Representa mediante clases una versión simplificada de un sistema de préstamos.

🕒 **Tiempo estimado:** 25 minutos

🔴 **Dificultad:** Difícil

Implementa:

### `Libro`

Atributos:

- `isbn`
- `titulo`
- `disponible`, inicialmente `True`

### `Usuario`

Atributos:

- `identificador`
- `nombre`
- `prestamos`, inicialmente una lista vacía

Métodos:

- `solicitar(libro)`: presta el libro si está disponible
- `devolver(libro)`: devuelve un libro prestado

Los métodos devolverán `True` cuando la operación se realice y `False` en caso contrario.

### 🧠 Antes de programar
- Al prestar un libro, debe dejar de estar disponible.
- El mismo libro no puede prestarse dos veces simultáneamente.
- Al devolverlo, debe eliminarse de la lista de préstamos y volver a estar disponible.

````python
class Libro:

    def __init__(self, isbn, titulo):
        self.isbn = isbn
        self.titulo = titulo
        self.disponible = True


class Usuario:

    def __init__(self, identificador, nombre):
        self.identificador = identificador
        self.nombre = nombre
        self.prestamos = []

    def solicitar(self, libro):
        if libro.disponible:
            libro.disponible = False
            self.prestamos.append(libro)
            return True
        return False

    def devolver(self, libro):
        if libro in self.prestamos:
            self.prestamos.remove(libro)
            libro.disponible = True
            return True
        return False
````

Ejecuta la siguiente celda para comprobar tu solución.

````python
libro = Libro("978-1", "Estructuras de Datos")
ana = Usuario("U01", "Ana")
luis = Usuario("U02", "Luis")

assert libro.disponible is True
assert ana.prestamos == []

assert ana.solicitar(libro) is True
assert libro.disponible is False
assert libro in ana.prestamos

assert luis.solicitar(libro) is False
assert libro not in luis.prestamos

assert ana.devolver(libro) is True
assert libro.disponible is True
assert libro not in ana.prestamos

assert ana.devolver(libro) is False
print("Pruebas superadas")
````
