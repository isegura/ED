# 2. Reto 1 — Diseña una clase `Alumno`

20–25 minutos


Diseña una clase `Alumno` que:

- almacene `nombre`, `id` y `notas`
- pueda calcular su media
- pueda decir si aprueba o no




````python
# Solución de referencia del reto 1

class Alumno:
    def __init__(self, nombre, id, notas):
        self.nombre = nombre
        self.id = id
        self.notas = notas

    def media(self):
        return sum(self.notas) / len(self.notas)

    def aprueba(self):
        return self.media() >= 5


a1 = Alumno("Ana", 1, [7, 8])
a2 = Alumno("Luis", 2, [5, 6])

print(a1.media(), a1.aprueba())
print(a2.media(), a2.aprueba())
````

````python
# Tests rápidos para validar en directo

a = Alumno("Elena", 10, [6, 7, 8])

assert a.nombre == "Elena"
assert a.id == 10
assert abs(a.media() - 7.0) < 1e-9
assert a.aprueba() is True

print("Reto 1 validado correctamente")
````

Diseña lo necesario para que:

- exista una clase `Asignatura`
- un objeto `Alumno` tenga varias asignaturas
- se pueda añadir una asignatura a un alumno
- se pueda calcular la media del alumno a partir de sus asignaturas


### Pista 1
¿Tiene sentido seguir usando solo una lista de números?

### Pista 2

### Pista 3
¿Puede un alumno guardar una lista de objetos `Asignatura`?

### Pista 4
¿Cómo cambia la media si ahora las notas están dentro de objetos?




````python
# Solución de referencia del reto 2

class Asignatura:
    def __init__(self, nombre, nota):
        self.nombre = nombre
        self.nota = nota


class Alumno:
    def __init__(self, nombre, id):
        self.nombre = nombre
        self.id = id
        self.asignaturas = []

    def anadir_asignatura(self, asignatura):
        self.asignaturas.append(asignatura)

    def media(self):
        return sum(asig.nota for asig in self.asignaturas) / len(self.asignaturas)


a = Alumno("Marta", 3)
a.anadir_asignatura(Asignatura("ED", 8))
a.anadir_asignatura(Asignatura("Matemática Discreta", 7))
a.anadir_asignatura(Asignatura("Programación", 9))

print(a.media())
````

````python
# Tests rápidos del reto 2

a = Alumno("Carlos", 20)
a.anadir_asignatura(Asignatura("ED", 5))
a.anadir_asignatura(Asignatura("POO", 7))

assert a.nombre == "Carlos"
assert len(a.asignaturas) == 2
assert a.asignaturas[0].nombre == "ED"
assert abs(a.media() - 6.0) < 1e-9

print("Reto 2 validado correctamente")
````

# Reto 3 — Mostrar objetovs

Cuando hacemos `print(objeto)` con una clase propia, ¿qué se muestra?


Modifica la clase `Alumno` para que, al ejecutar:

```python
print(alumno)
```

se muestre una salida clara, por ejemplo:

```python
Ana (ID: 1) - Media: 8.50
```



````python
# Solución de referencia del reto 3

class Asignatura:
    def __init__(self, nombre, nota):
        self.nombre = nombre
        self.nota = nota


class Alumno:
    def __init__(self, nombre, id):
        self.nombre = nombre
        self.id = id
        self.asignaturas = []

    def anadir_asignatura(self, asignatura):
        self.asignaturas.append(asignatura)

    def media(self):
        return sum(asig.nota for asig in self.asignaturas) / len(self.asignaturas)

    def __str__(self):
        return f"{self.nombre} (ID: {self.id}) - Media: {self.media():.2f}"


a = Alumno("Ana", 1)
a.anadir_asignatura(Asignatura("ED", 8))
a.anadir_asignatura(Asignatura("Programación", 9))

print(a)
````

````python
# Test rápido del reto 3

a = Alumno("Nora", 99)
a.anadir_asignatura(Asignatura("ED", 6))
a.anadir_asignatura(Asignatura("Python", 8))

texto = str(a)
assert "Nora" in texto
assert "99" in texto
assert "7.00" in texto

print("Reto 3 validado correctamente")
````

# Reto 7 - Añadir una función que permita buscar un alumno por id

Implementa una función que reciba una lista de alumnos y un id, y busque el alumno con ese id. La función debe mostrar al alumno. Si el alumno no existe, la función debe informar que no existe.



````python
# Bonus A: búsqueda lineal por id

def buscar_alumno_por_id(lista_alumnos, id_buscado):
    for alumno in lista_alumnos:
        if alumno.id == id_buscado:
            return alumno
    return None


a1 = Alumno("Ana", 1)
a1.anadir_asignatura(Asignatura("ED", 8))
a1.anadir_asignatura(Asignatura("Prog", 9))

a2 = Alumno("Luis", 2)
a2.anadir_asignatura(Asignatura("ED", 5))
a2.anadir_asignatura(Asignatura("Prog", 6))

grupo = [a1, a2]

encontrado = buscar_alumno_por_id(grupo, 2)
print(encontrado)
````
