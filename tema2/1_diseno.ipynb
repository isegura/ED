## ¿Qué es diseñar un algoritmo?

Diseñar un algoritmo es **pensar cómo resolver un problema paso a paso antes de escribir código**.

Pasos:
1. Entender el problema  
2. Pensar la solución  
3. Escribir los pasos  
4. Programarlo  
5. Probarlo

## Ejemplo guiado

Problema: encontrar el número mayor de una lista

### Pensamiento:
- recorrer la lista  
- comparar valores  
- guardar el mayor  

### Código:

```python
numeros = [3, 8, 2, 10, 5]

mayor = numeros[0]

for n in numeros:
    if n > mayor:
        mayor = n

print("Mayor:", mayor)
```

## 🟢 Retos básicos

### Reto 1 - Sumar elementos

Dada una lista de números, calcula la suma total.

---

### Reto 2 - Contar elementos

Cuenta cuántos números hay en una lista sin usar `len()`.

---

### Reto 3 - Contar mayores que 10

Cuenta cuántos números de una lista son mayores que 10.

---

```python
# Soluciones básicas

# Reto 1
numeros = [1, 2, 3, 4]
suma = 0
for n in numeros:
    suma += n
print("Suma:", suma)

# Reto 2
contador = 0
for _ in numeros:
    contador += 1
print("Cantidad:", contador)

# Reto 3
contador = 0
for n in numeros:
    if n > 10:
        contador += 1
print("Mayores que 10:", contador)
```

## 🟡 Retos intermedios

### Reto 4 - Encontrar el menor

Encuentra el número más pequeño de una lista.

---

### Reto 5 - Filtrar pares

Crea una nueva lista solo con números pares.

---

### Reto 6 - Buscar un elemento

Comprueba si un número está en la lista.

---

```python
# Soluciones intermedias

# Reto 4
numeros = [5, 2, 8, 1]
menor = numeros[0]
for n in numeros:
    if n < menor:
        menor = n
print("Menor:", menor)

# Reto 5
pares = []
for n in numeros:
    if n % 2 == 0:
        pares.append(n)
print("Pares:", pares)

# Reto 6
objetivo = 8
encontrado = False
for n in numeros:
    if n == objetivo:
        encontrado = True
print("¿Está?", encontrado)
```

## 🔴 Retos avanzados

### Reto 7 - Lista sin duplicados

Dada una lista, crea otra sin elementos repetidos (sin usar `set`).

---

### Reto 8 - Segundo mayor

Encuentra el segundo número más grande.

---

### Reto 9 - Contar palabras

Dada una lista de palabras, cuenta cuántas tienen más de 5 letras.

---

```python
# Soluciones avanzadas

# Reto 7
numeros = [1, 2, 2, 3, 4, 4]
sin_repetidos = []

for n in numeros:
    if n not in sin_repetidos:
        sin_repetidos.append(n)

print(sin_repetidos)

# Reto 8
numeros = [10, 5, 8, 20]
mayor = max(numeros)
segundo = None

for n in numeros:
    if n != mayor:
        if segundo is None or n > segundo:
            segundo = n

print("Segundo mayor:", segundo)

# Reto 9
palabras = ["python", "sol", "programacion"]
contador = 0

for p in palabras:
    if len(p) > 5:
        contador += 1

print("Palabras largas:", contador)
```

## 🧠 Reto final

Crea un programa que:

- reciba una lista de números  
- elimine los repetidos  
- calcule la media  
- muestre el mayor y el menor  

---

```python
# Solución reto final

numeros = [1, 2, 2, 3, 4, 4]

sin_repetidos = []
for n in numeros:
    if n not in sin_repetidos:
        sin_repetidos.append(n)

suma = 0
for n in sin_repetidos:
    suma += n

media = suma / len(sin_repetidos)

print("Lista:", sin_repetidos)
print("Media:", media)
print("Mayor:", max(sin_repetidos))
print("Menor:", min(sin_repetidos))
```
