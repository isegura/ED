# 🔎 Ordenación y Búsqueda en Python

Aprenderás a ordenar datos y a buscarlos usando algoritmos básicos.


## 🔄 Ordenación

Ordenar consiste en organizar datos (normalmente números o palabras).


```python
numeros = [5, 2, 9, 1]

print("Ordenados:", sorted(numeros))
print("Original:", numeros)
```

### Algoritmo: Burbuja

Compara elementos vecinos e intercambia si están en mal orden.


```python
numeros = [5, 2, 9, 1]

for i in range(len(numeros)):
    for j in range(len(numeros)-1):
        if numeros[j] > numeros[j+1]:
            numeros[j], numeros[j+1] = numeros[j+1], numeros[j]

print(numeros)
```

## 🔍 Búsqueda

Buscar consiste en encontrar un elemento dentro de una colección.


```python
numeros = [3, 8, 2, 10]

objetivo = 8
encontrado = False

for n in numeros:
    if n == objetivo:
        encontrado = True

print("¿Encontrado?", encontrado)
```

### Búsqueda binaria (lista ordenada)


```python
numeros = [1, 3, 5, 7, 9]

objetivo = 7
inicio = 0
fin = len(numeros) - 1

while inicio <= fin:
    medio = (inicio + fin) // 2

    if numeros[medio] == objetivo:
        print("Encontrado")
        break
    elif numeros[medio] < objetivo:
        inicio = medio + 1
    else:
        fin = medio - 1
```
