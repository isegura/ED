# Análisis empírico de algoritmos


En este ejemplo se muestra cómo realizar un análisis empírico de un algoritmo, midiendo su tiempo de ejecución para distintos tamaños de entrada.

**Pasos del análisis empírico**

1. Implementar el algoritmo.
2. Medir el tiempo de ejecución.
3. Ejecutarlo con distintos tamaños de entrada.
4. Registrar los tiempos obtenidos.
5. Representar Tiempo vs. Tamaño de entrada.
   mediante una gráfica.


## 1) Solución iterativa

Implementar una función que calcule la suma de los primeros N números naturales utilizando un algoritmo iterativo.

```python
import time
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sumar_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
```

Ahora medimos el tiempo de ejecución para distintos tamaños de entrada


```python
mediciones = []

for i in range(1, 8):

    n = 10**i

    inicio = time.time()
    resultado = sumar_n(n)
    fin = time.time()

    tiempo = fin - inicio

    mediciones.append([n, tiempo])

df = pd.DataFrame(mediciones, columns=["N", "Tiempo (s)"])
df
```

El siguiente paso es obtener la representación gráfica que permita estudiar cómo se comporta el tiempo de ejecución en función del tamaño de entrada.

```python
plt.figure(figsize=(8,5))

plt.plot(
    df["N"],
    df["Tiempo (s)"],
    marker="o",
    linewidth=2
)

plt.title("Análisis empírico del algoritmo (solución iterativa)")
plt.xlabel("Tamaño de entrada (N)")
plt.ylabel("Tiempo de ejecución (s)")
plt.grid(True)

plt.show()
```

## 2) Solución basada en la suma de Gauss



```python
def sumar_gauss(n):
    return n * (n + 1) // 2

# Medición del tiempo de ejecución (O(1))
mediciones = []

for i in range(1, 8):

    n = 10**i

    inicio = time.time()
    resultado = sumar_gauss(n)
    fin = time.time()

    tiempo = fin - inicio

    mediciones.append([n, tiempo])

df_gauss = pd.DataFrame(mediciones, columns=["N", "Tiempo (s)"])
print(df_gauss)

# Representación gráfica

plt.figure(figsize=(8,5))

plt.plot(
    df_gauss["N"],
    df_gauss["Tiempo (s)"],
    marker="o",
    linewidth=2
)


plt.title("Análisis empírico del algoritmo (suma de Gauss)")
plt.xlabel("Tamaño de entrada (N)")
plt.ylabel("Tiempo de ejecución (s)")
plt.grid(True)

plt.show()
```

## Comparación: iterativa vs gauss
Se utiliza una escala logarítmica en el eje Y para facilitar la comparación entre algoritmos con tiempos de ejecución muy diferentes.


```python
plt.figure(figsize=(8,5))

plt.plot(df["N"],
         df["Tiempo (s)"],
         marker="o",
         linewidth=2,
         label="Suma iterativa")

plt.plot(df_gauss["N"],
         df_gauss["Tiempo (s)"],
         marker="s",
         linewidth=2,
         label="Suma de Gauss")

# Utiliza una escala logarítmica en el eje Y para facilitar la comparación
# entre algoritmos con tiempos de ejecución muy diferentes.
plt.yscale("log")

plt.title("Comparación del tiempo de ejecución")
plt.xlabel("Tamaño de entrada (N)")
plt.ylabel("Tiempo de ejecución (s)")
plt.grid(True)
plt.legend()

plt.show()
```
