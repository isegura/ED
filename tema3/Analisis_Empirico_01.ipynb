# Análisis empírico de algoritmos

## Objetivo

En este ejemplo se muestra cómo realizar un análisis empírico de un algoritmo, midiendo su tiempo de ejecución para distintos tamaños de entrada.

**Pasos del análisis empírico**

1. Implementar el algoritmo.
2. Medir el tiempo de ejecución.
3. Ejecutarlo con distintos tamaños de entrada.
4. Registrar los tiempos obtenidos.
5. Representar Tiempo vs. Tamaño de entrada.
   mediante una gráfica.


### Implementación del algoritmo

Implementar una función que calcule la suma de los primeros N números naturales utilizando un algoritmo iterativo.

```python
import time
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sumar_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
```

### Medición del tiempo de ejecución

```python
mediciones = []

for i in range(1, 8):

    n = 10**i

    inicio = time.time()
    resultado = sumar_n(n)
    fin = time.time()

    tiempo = fin - inicio

    mediciones.append([n, tiempo])

df = pd.DataFrame(mediciones, columns=["N", "Tiempo (s)"])
df
```

### Representación gráfica

```python
plt.figure(figsize=(8,5))

plt.plot(
    df["N"],
    df["Tiempo (s)"],
    marker="o",
    linewidth=2
)

plt.title("Análisis empírico del algoritmo")
plt.xlabel("Tamaño de entrada (N)")
plt.ylabel("Tiempo de ejecución (s)")
plt.grid(True)

plt.show()
```

### Conclusiones
- La representación gráfica permite observar la tendencia de crecimiento del algoritmo.
- El tiempo de ejecución aumenta al incrementar el tamaño de la entrada.
- El análisis empírico proporciona una estimación experimental del rendimiento
