# Uso de `assert` en Python

## Objetivo

En este notebook aprenderás qué es la instrucción `assert`, cómo funciona y cómo utilizarla para comprobar automáticamente que una función devuelve el resultado esperado.


## ¿Qué es `assert`?

La sintaxis es:

```python
assert condición
```

- Si la condición es **verdadera**, el programa continúa.
- Si la condición es **falsa**, Python lanza un **`AssertionError`**.


````
# Ejemplos sencillos

assert 2 + 3 == 5
assert 10 > 5
assert "hola".upper() == "HOLA"

print("Todas las comprobaciones son correctas.")
````

## ¿Qué ocurre cuando la condición es falsa?

En la siguiente celda se produce un `AssertionError` porque el resultado obtenido no coincide con el esperado.


````
def cuadrado_v1(x):
    return x * x + 1

# Esta celda lanzará una excepción
assert cuadrado_v1(4) == 16
````

### ¿Por qué se produce el error?

La función devuelve `17`, pero el `assert` espera `16`.

Es decir, Python comprueba:

```python
17 == 16
```

Como esta condición es falsa, se produce un **`AssertionError`**.


## Uso de `assert` para probar funciones

Una vez implementada una función, podemos escribir varias pruebas para verificar automáticamente que funciona correctamente.


````
def cuadrado(x):
    return x * x

assert cuadrado(0) == 0      # Caso sencillo
assert cuadrado(4) == 16     # Número positivo
assert cuadrado(-3) == 9     # Número negativo

print("Pruebas superadas.")
````

## Resumen

- `assert` permite comprobar automáticamente que una condición es cierta.
- Si la condición se cumple, el programa continúa.
- Si no se cumple, Python lanza un `AssertionError`.
- Es una herramienta muy útil para verificar que nuestras funciones producen los resultados esperados.
