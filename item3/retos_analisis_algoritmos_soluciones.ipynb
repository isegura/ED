# 🏆 Reto 1. Contar signos

🎯 **Objetivo:** Analiza la complejidad temporal y espacial de la siguiente función.

🕒 **Tiempo estimado:** 8 minutos  
🟢 **Dificultad:** Fácil


```python
def resumen_signos(valores):
    resumen = {
        "positivos": 0,
        "negativos": 0,
        "ceros": 0,
    }

    for valor in valores:
        if valor > 0:
            resumen["positivos"] += 1
        elif valor < 0:
            resumen["negativos"] += 1
        else:
            resumen["ceros"] += 1

    return resumen
```

### 🔍 Preguntas

1. ¿Qué representa \(n\) en este algoritmo?
2. ¿Cuántas veces se ejecuta el bucle?
3. ¿Cuál es su complejidad temporal?
4. ¿Cuál es su complejidad espacial auxiliar?
5. ¿Cambian el mejor y el peor caso?

### ✍️ Respuesta

- **\(n\):** representa el número de elementos de la lista `valores`.
- **Número de iteraciones:** el bucle se ejecuta \(n\) veces.
- **Complejidad temporal:** \(O(n)\).
- **Complejidad espacial auxiliar:** \(O(1)\).
- **Mejor y peor caso:** no cambian; en ambos casos se recorren todos los elementos.

**Justificación:** en cada iteración se realiza un número constante de operaciones. El diccionario `resumen` siempre contiene tres entradas, independientemente del tamaño de la lista.

# 🏆 Reto 2. Totales acumulados

🎯 **Objetivo:** Analiza la complejidad temporal y espacial de la siguiente función.

🕒 **Tiempo estimado:** 8 minutos  
🟢 **Dificultad:** Fácil


```python
def acumulados(valores):
    resultado = []
    total = 0

    for valor in valores:
        total += valor
        resultado.append(total)

    return resultado
```

### 🔍 Preguntas

1. ¿Cuántos elementos contiene `resultado` al finalizar?
2. ¿Cuál es la complejidad temporal?
3. ¿Cuál es la complejidad espacial?
4. ¿Podría reducirse la complejidad espacial sin modificar lo que devuelve la función?

### ✍️ Respuesta

- **Elementos de `resultado`:** \(n\), uno por cada elemento de `valores`.
- **Complejidad temporal:** \(O(n)\).
- **Complejidad espacial:** \(O(n)\).
- **¿Puede reducirse?:** no, si la función debe seguir devolviendo una lista con todos los acumulados.

**Justificación:** el algoritmo recorre la entrada una sola vez. La lista devuelta crece hasta contener \(n\) elementos, por lo que necesita espacio lineal.

# 🏆 Reto 3. Eliminar duplicados

🎯 **Objetivo:** Analiza la complejidad temporal y espacial de la siguiente función:

🕒 **Tiempo estimado:** 12 minutos  
🟡 **Dificultad:** Media


```python
def sin_duplicados(valores):
    resultado = []

    for valor in valores:
        if valor not in resultado:
            resultado.append(valor)

    return resultado
```

### 🔍 Preguntas

1. ¿Cuál es el coste de `valor not in resultado` cuando `resultado` es una lista?
2. ¿De qué depende el número de comparaciones que realiza esa operación?
3. ¿Cuál es el peor caso para este algoritmo? Justifica tu respuesta.
4. ¿Cuál es la complejidad temporal del algoritmo?
5. ¿Cuál es la complejidad espacial del algoritmo?
6. ¿Qué estructura de datos de Python utilizarías para reducir el coste de las comprobaciones de pertenencia? ¿Cómo cambiaría la complejidad temporal del algoritmo?

### ✍️ Respuesta

- **Coste de `valor not in resultado`:** \(O(k)\), donde \(k\) es el número de elementos que contiene `resultado` en ese momento.
- **Número de comparaciones:** depende del tamaño de `resultado` y de la posición en la que se encuentre el elemento, si ya estaba incluido.
- **Peor caso:** todos los valores son diferentes. En ese caso, `resultado` crece en cada iteración y se realizan aproximadamente \(0 + 1 + 2 + \cdots + (n-1)\) comparaciones.
- **Complejidad temporal:** \(O(n^2)\) en el peor caso.
- **Complejidad espacial:** \(O(n)\).
- **Posible mejora:** utilizar un `set` para comprobar si un elemento ya ha aparecido. La pertenencia en un conjunto tiene coste medio \(O(1)\), por lo que el algoritmo completo pasaría a tener un coste temporal medio \(O(n)\), manteniendo \(O(n)\) de espacio.

# 🏆 Reto 4. Sumas por filas

🎯 **Objetivo:** Analiza un algoritmo que recorre una matriz rectangular (tiene dimensión f x c) mediante bucles anidados para calcular la suma de los elementos de cada fila.

🕒 **Tiempo estimado:** 12 minutos  
🟡 **Dificultad:** Media

La siguiente función recibe una **matriz** representada como una lista de listas de números enteros, donde todas las filas tienen el mismo número de elementos. La función devuelve una nueva lista con la suma de cada fila.

```python
def sumas_filas(matriz):
    resultado = []

    for fila in matriz:
        total_fila = 0

        for valor in fila:
            total_fila += valor

        resultado.append(total_fila)

    return resultado
```

### 🔍 Preguntas

Sea \(f\) el número de filas y \(c\) el número de columnas de la matriz.

1. ¿Cuántas veces se ejecuta la instrucción `suma += valor`?
2. Expresa la complejidad temporal del algoritmo en función de \(f\) y \(c\).
3. Si la matriz es cuadrada de tamaño \(n \times n\), ¿cuál es la complejidad temporal?
4. ¿Cuál es la complejidad espacial de la lista `resultado`?

### ✍️ Respuesta

- **Número de ejecuciones de `total_fila += valor`:**  
  Se ejecuta una vez por cada elemento de la matriz, es decir, \(f \cdot c\) veces.

- **Complejidad temporal para una matriz \(f \times c\):**  
  \[
  O(f \cdot c)
  \]

- **Complejidad temporal para una matriz \(n \times n\):**  
  \[
  O(n^2)
  \]

- **Complejidad espacial:**  
  La lista `resultado` almacena una suma por cada fila, por lo que ocupa \(O(f)\).

# 🏆 Reto 5. Validar una matriz rectangular

🎯 **Objetivo**: Analiza cómo cambia el tiempo de ejecución de un algoritmo según el caso de entrada.
🕒 **Tiempo estimado:** 10 minutos  
🟡 **Dificultad:** Media


```python
def es_rectangular(matriz):
    if len(matriz) == 0:
        return True

    longitud = len(matriz[0])

    for fila in matriz:
        if len(fila) != longitud:
            return False

    return True
```

### 🔍 Preguntas

Sea \(n\) el número de filas.

1. ¿Qué entrada produce el mejor caso?
2. ¿Qué entrada produce el peor caso?
3. ¿Cuál es la complejidad temporal del mejor caso?
4. ¿Cuál es la complejidad temporal del peor caso?
5. ¿Cuál es la complejidad espacial auxiliar?

### ✍️ Respuesta

- **Mejor caso:** la matriz está vacía, o bien la primera fila que se comprueba tiene una longitud distinta de la primera fila. En ambos casos la función termina inmediatamente o tras un número constante de comprobaciones.
- **Peor caso:** todas las filas tienen la misma longitud, o la primera fila distinta aparece al final. Es necesario examinar las \(n\) filas.
- **Complejidad temporal del mejor caso:** \(O(1)\).
- **Complejidad temporal del peor caso:** \(O(n)\).
- **Complejidad espacial auxiliar:** \(O(1)\).

**Justificación:** `len(fila)` tiene coste constante en una lista de Python y el algoritmo no crea ninguna estructura cuyo tamaño dependa de \(n\).

# 🏆 Reto 6. Coste constante frente a coste lineal

🎯 **Objetivo:** Compara dos algoritmos que resuelven el mismo problema.

🕒 **Tiempo estimado:** 10 minutos  
🟢 **Dificultad:** Fácil


```python
def sumar_iterativa(n):
    total = 0

    for i in range(1, n + 1):
        total += i

    return total


def sumar_gauss(n):
    return n * (n + 1) // 2
```

### 🔍 Preguntas

1. ¿Cuál es la complejidad temporal de cada función?
2. ¿Cuál es la complejidad espacial de cada función?
3. ¿Cuál escala mejor cuando \(n\) es muy grande?
4. ¿Qué función utilizarías para \(n = 10^9\)? Justifica la respuesta.

### ✍️ Respuesta

| Función | Complejidad temporal | Complejidad espacial |
|---|---|---|
| `sumar_iterativa` | \(O(n)\) | \(O(1)\) |
| `sumar_gauss` | \(O(1)\) | \(O(1)\) |

**Conclusión:** `sumar_gauss` escala mejor porque realiza un número constante de operaciones, independientemente del valor de \(n\). Para \(n = 10^9\), utilizaría `sumar_gauss`, ya que evita ejecutar mil millones de iteraciones.

# 🧪 Reto 7. Análisis empírico

🎯 **Objetivo:** Mide experimentalmente el tiempo de ejecución de un algoritmo para distintos tamaños de entrada y representa los resultados.

🕒 **Tiempo estimado:** 20 minutos  
🟡 **Dificultad:** Media


```python
def contar_pares(valores):
    cantidad = 0

    for valor in valores:
        if valor % 2 == 0:
            cantidad += 1

    return cantidad
```

### Tareas

1. Genera listas con tamaños \(10^1, 10^2, \ldots, 10^7\).
2. Mide el tiempo de ejecución con `time.time()`.
3. Guarda cada par `(tamaño, tiempo)` en una lista llamada `mediciones`.
4. Crea un `DataFrame` con los resultados.
5. Representa una gráfica de **Tiempo de ejecución vs. Tamaño de entrada**.
6. Indica qué tendencia observas y compárala con el análisis asintótico.


```python
import time
import pandas as pd
import matplotlib.pyplot as plt

mediciones = []

for exponente in range(1, 8):
    n = 10 ** exponente
    valores = list(range(n))

    inicio = time.time()
    contar_pares(valores)
    fin = time.time()

    mediciones.append((n, fin - inicio))

df = pd.DataFrame(mediciones, columns=["N", "Tiempo (s)"])
df
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df["N"], df["Tiempo (s)"], marker="o")

plt.title("Análisis empírico de contar_pares")
plt.xlabel("Tamaño de entrada (N)")
plt.ylabel("Tiempo de ejecución (s)")
plt.grid(True)
plt.show()
```

### ✍️ Conclusión

- **Tendencia observada:** el tiempo de ejecución aumenta aproximadamente de forma lineal al crecer el tamaño de la lista.
- **Complejidad temporal esperada:** \(O(n)\).
- **¿Coinciden el análisis empírico y el asintótico?:** sí. La función recorre todos los elementos una sola vez y realiza una cantidad constante de trabajo por elemento. Las mediciones pueden presentar pequeñas variaciones debidas al sistema y al propio mecanismo de medida.
